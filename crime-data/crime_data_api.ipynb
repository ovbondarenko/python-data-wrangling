```python
with open('data/agencies.txt') as json_data:
    d = json.load(json_data)

agencies_df = pd.DataFrame(d)
# agencies_df[2195:]
```

```python
def collect_agency_report(ori):

    try:
        query_url = f'https://api.usa.gov/crime/fbi/sapi/api/summarized/agencies/{ori}/offenses/2014/2017?api_key={ckey}'
        response = requests.get(query_url, timeout = 30)
        print('==========================================')
        print(f'Requesting report from agency {ori}')
        print(f'Time: {datetime.datetime.now()}')
        result = response.json()
        report_list = result['results']
    except:
        report_list = [{'ori': ori}]
        print('............................................')
        print(f'Report not found, skipping agency {ori}...')
    return report_list
```
